# 5) Let AVOCADO  Average Weight : 
    
    4046 => 4 ounces, 4225 => 9 ounces, 4770 => 12 ounces
    1. Find the number of sold avocadoes by region ?

```python
#Group region หลังจากนั้นหาผลรวมตาม region
sumvol5 = avocado.groupby('region').sum()

#Drop the region that are not the city name
sumvol5 = sumvol5.drop(['TotalUS', 'Midsouth', 'Northeast', 'West', 'Southeast', 'SouthCentral', 'Plains'])

#จะหาค่าเฉลี่ยในหน่วย ounce ดังนั้นเปลี่ยนจาก units เป็น ounce ในส่วนของ columns 4046, 4225,4770 ด้วยการหาร 4, 9 และ 12
sumvol5['4046'] = sumvol5['4046']/4
sumvol5['4225'] = sumvol5['4225']/9
sumvol5['4770'] = sumvol5['4770']/12

#สร้าง column ใหม่ คือ จำนวนอะโวคาโดที่ขายตามภูมิภาค โดยใช้ round() ควบคุมให้เป็นเลขจำนวนเต็ม
sumvol5['Total_avocadoes_by_region'] = round(sumvol5['4046'] + sumvol5['4225'] + sumvol5['4770'])


print('The number of sold avocadoes by region as this following')
sumvol5
```

Which region sold the largest number of avocados ?

```python
q5 = sumvol5.loc[sumvol5['Total_avocadoes_by_region'].idxmax()]
q5ans = sumvol5['Total_avocadoes_by_region'].idxmax()
q5max = sumvol5.loc[sumvol5['Total_avocadoes_by_region'].idxmax()][11] # ให้ใช้ [11] เลือก column ที่จะแสดง โดย Total_avocadoes_by_region อยู่ column ที่ 11 
q5
```

```python
print(f'The region that sold largest avocado is "{q5ans}" with total avocados sold equal {round(q5max,2)} units.')
```
